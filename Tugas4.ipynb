```python
import numpy as np

# --- FUNGSI DEKOMPOSISI LU (MANUAL) ---
def lu_decomposition(A):
    n = len(A)
    L = np.eye(n)
    U = A.copy().astype(float)

    for i in range(n):
        for j in range(i + 1, n):
            factor = U[j, i] / U[i, i]
            L[j, i] = factor
            U[j, i:] -= factor * U[i, i:]
    return L, U

def solve_lu(L, U, v):
    n = len(L)
    # Forward Substitution (Ly = v)
    y = np.zeros(n)
    for i in range(n):
        y[i] = v[i] - np.dot(L[i, :i], y[:i])
    # Backward Substitution (Ux = y)
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (y[i] - np.dot(U[i, i+1:], x[i+1:])) / U[i, i]
    return x

# --- FUNGSI JACOBI ---
def solve_jacobi(A, B, x, tol=1e-10, max_iter=100):
    n = len(B)
    x_new = np.zeros_like(x)
    for k in range(max_iter):
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x_new[i] = (B[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new.copy()
    return x, max_iter

# --- FUNGSI GAUSS-SEIDEL ---
def solve_gauss_seidel(A, B, x, tol=1e-10, max_iter=100):
    n = len(B)
    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x[i] = (B[i] - s) / A[i][i]
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            return x, k
    return x, max_iter
```

```python
# Definisi Matriks A dan Vektor B
A = np.array([[10.0, -2.0, -3.0, 0.0],
              [-2.0, 12.0, -4.0, -1.0],
              [-3.0, -4.0, 15.0, -2.0],
              [0.0, -1.0, -2.0, 8.0]])

B = np.array([10.0, -5.0, 0.0, 15.0])
x_init = np.zeros(len(B))

# 1. Penyelesaian LU
L, U = lu_decomposition(A)
res_lu = solve_lu(L, U, B)

# 2. Penyelesaian Jacobi
res_jacobi, iter_j = solve_jacobi(A, B, x_init.copy())

# 3. Penyelesaian Gauss-Seidel
res_gs, iter_gs = solve_gauss_seidel(A, B, x_init.copy())

print("=== HASIL PERHITUNGAN ===")
print(f"Solusi LU           : {res_lu}")
print(f"Solusi Jacobi       : {res_jacobi} (Iterasi: {iter_j})")
print(f"Solusi Gauss-Seidel : {res_gs} (Iterasi: {iter_gs})")
```
