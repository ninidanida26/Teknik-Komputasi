```python
sentence = "this is a rambling sentence that simply goes on and on and on and just simply will not stop that is just the way things are"


words = sentence.split()


wordtocount = {}


for word in words:
    wordtocount[word] = wordtocount.get(word, 0) + 1


print("Word\t\tCount")
for word, count in wordtocount.items():
    print(f"{word}\t\t{count}")
```

```python
def jumlah_ganjil(n):
    total = 0
    for i in range(1, n + 1, 2):
        total += i
    return total


while True:
    angka = int(input("Masukkan bilangan ganjil besar (genap untuk berhenti): "))

    if angka % 2 == 0:
        print("Bilangan genap dimasukkan. Program berhenti.")
        break
    else:
        hasil = jumlah_ganjil(angka)
        print("Jumlah bilangan ganjil sampai", angka, "adalah", hasil)
```

```python
def sum_of_odds(n):
    total = 0
    for i in range(1, n + 1, 2):
        total += i
    return total

with open("odd_in.dat", "r") as infile, open("odd_out.dat", "w") as outfile:
    counter = 1
    for line in infile:
        line = line.strip()

        if not line:
            continue

        try:
            n = int(line.split()[0])

            if n > 0 and n % 2 != 0:
                result = sum_of_odds(n)
                outfile.write(f"{counter} {n} {result}\n")
                print(counter, n, result)
                counter += 1

        except ValueError:
            continue
```

```python
import math
import os

if not os.path.exists("dummy1.dat"):
    with open("dummy1.dat", "w") as f:
        f.write("Halo dunia 21.2 12.1\n")
        f.write("bagaimana 1.7 123.5\n")
        f.write("Anda mengaburkan 0,004 19,5\n")

total_sum = 0.0
counter = 1

with open("dummy1.dat", "r") as infile, open("dummy2.dat", "w") as outfile:
    for line in infile:
        line = line.strip()
        if not line:
            continue

        parts = line.split()

        try:
            num1 = float(parts[-2].replace(",", "."))
            num2 = float(parts[-1].replace(",", "."))

            if num2 != 0:
                exponent = int(math.floor(math.log10(abs(num2))))
            else:
                exponent = 0


            outfile.write(f"{counter} {num2} {exponent}\n")


            print(counter, num2, exponent)

            total_sum += num1 + num2
            counter += 1

        except ValueError:
            continue

    outfile.write(f"{total_sum}\n")
    print(total_sum)
```

```python
import os

def readtable(filename):
    x = []
    y = []
    with open(filename) as f:
        for line in f:
            if line.strip():
                xi, yi = line.split()
                x.append(float(xi))
                y.append(float(yi))
    return x, y


if not os.path.exists("data.dat"):
    with open("data.dat", "w") as f:
        f.write("1 2\n")
        f.write("2 4\n")
        f.write("3 6\n")

x, y = readtable("data.dat")
print("x =", x)
print("y =", y)
```

```python
import math

def f(x):
    return math.exp(-x**4)

x_values = [i / 10 for i in range(-10, 11)]

f_values = [f(x) for x in x_values]

print("   x      f(x)")
print("-----------------")

for x, fx in zip(x_values, f_values):
    print(f"{x:.1f}    {fx:.5f}")
```
